```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 1: Generate 20 data points in different ways
np.random.seed(42)  # Setting seed for reproducibility

# Case 1: Linear relationship
x1 = np.linspace(0, 10, 20)
y1 = 2 * x1 + 1 + np.random.normal(scale=2, size=20)

# Case 2: Quadratic relationship
x2 = np.linspace(-5, 5, 20)
y2 = 1.5 * x2**2 + 3 * x2 + 2 + np.random.normal(scale=5, size=20)

# Case 3: Noisy random data
x3 = np.random.rand(20) * 10
y3 = np.random.rand(20) * 5 + 2 * x3 + 1 + np.random.normal(scale=2, size=20)
```

```python
plt.scatter(x1,y1)
```

```python
plt.scatter(x2,y2)
```

```python
plt.scatter(x3,y3)
```

```python
# Step 2: Implement the gradient descent method for linear regression
def linear_regression_gradient_descent(x, y, learning_rate=0.01, epochs=1000):
    m, b = 0, 0  # Initial parameters
    n = len(x)

    for epoch in range(epochs):
        # Calculate predictions and errors
        y_pred = m * x + b
        error = y_pred - y

        # Update parameters using gradient descent
        m -= learning_rate * (2/n) * np.dot(error, x)
        b -= learning_rate * (2/n) * np.sum(error)

    return m, b
```

```python
# Step 3: Execute linear regression on the generated data
m_best1, b_best1 = linear_regression_gradient_descent(x1, y1)
```

```python
# Step 4: Plot the results
plt.scatter(x1, y1, label='Data points from Case 1')
plt.plot(x1, m_best1 * x1 + b_best1, color='red', label='Best Linear Fit')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Best Linear Fit using Gradient Descent')
plt.legend()
```

```python
# Step 5: Report the results
report = """
Results:
- Best Fit Line: y = {:.2f}x + {:.2f}
"""

print(report.format(m_best1, b_best1))
```
